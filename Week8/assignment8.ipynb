# Kmeans-Clustering & PCA <br>
In this assignment we will start implementing K-means algorithm which is a clustering algorithm gather similar data into groups based on K a number that is given to the algorithm.<br>
Then we will implement PCA (Principal Component analysis) which is a dimensionality reduction algorithm.

Now let's load our packages as usual then the data

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import matplotlib.image as mpimg
from numpy.linalg import svd
```

```python
mat = loadmat("./Data/ex7data2.mat")
X = mat["X"]
```

Now we will impelement find closest centroids funcation<br>
Input: X -> features<br>
&emsp; &emsp; &emsp; centroids -> list of centroids for different clusters<br>



Output: the closest centroid to the feauture X<br>

```python
def findClosestCentroids(X, centroids):
    K = centroids.shape[0]
    idx = np.zeros((X.shape[0], 1))
    temp = np.zeros((centroids.shape[0], 1))

    for i in range(X.shape[0]):
        for j in range(K):
            dist = X[i, :] - centroids[j,:]
            length = np.sum(dist**2)
            temp[j] = length
        idx[i] = np.argmin(temp)+1
    return idx
```

Now let's test what we have done so far

```python
K = 3
initial_centroids = np.array([[3,3],[6,2],[8,5]])
idx = findClosestCentroids(X, initial_centroids)
print("Closest centroids for the first 3 examples:\n",idx[0:3])
```

Now we will impelement a function that calculate the mean of the cluster and return the new centroids

```python
def computeCentroids(X, idx, K):
    m, n = X.shape[0], X.shape[1]
    centroids = np.zeros((K, n))
    count = np.zeros((K, 1))

    for i in range(m):
        index = int((idx[i]-1)[0])
        centroids[index,:]+=X[i,:]
        count[index]+=1
    return centroids/count
```

Now let's test computeCentroids funcation

```python
centroids = computeCentroids(X, idx, K)
print("Centroids computed after initial finding of closest centroids:\n", centroids)
```

### Now we will impelement a funcation that will help us visualizing K-means algorithm to understand

```python
def plotKmeans(X, centroids, idx, K, num_iters):
    m, n = X.shape[0], X.shape[1]

    fig, ax = plt.subplots(nrows=num_iters, ncols=1, figsize=(6, 36))

    for i in range(num_iters):
        color = "rgb"
        for k in range(1, K+1):
            grp = (idx==k).reshape(m,1)
            ax[i].scatter(X[grp[:,0],0], X[grp[:,0],1], c=color[k-1], s=15)
        
        ax[i].scatter(centroids[:,0], centroids[:,1], s=120, marker="x", c="black", linewidth=3)
        title = "Iteration Number " + str(i)
        ax[i].set_title(title)
    

        centroids = computeCentroids(X, idx, K)

        idx = findClosestCentroids(X, centroids)

    plt.tight_layout()
```

```python
m, n = X.shape[0], X.shape[1]
plotKmeans(X, initial_centroids, idx, K, 10)
```

Now let's try random initialization with K-means

```python
def kMeansInitCentroids(X, K):
    m,n = X.shape[0], X.shape[1]
    centroids = np.zeros((K,n))
    
    for i in range(K):
        centroids[i] = X[np.random.randint(0,m+1),:]
        
    return centroids
```

```python
centroids = kMeansInitCentroids(X, K)
idx = findClosestCentroids(X, centroids)
plotKmeans(X, centroids,idx, K,10)
```

# PCA

Load the data then visualize it

```python
mat3 = loadmat("./Data/ex7data1.mat")
X3 = mat3["X"]
```

```python
plt.scatter(X3[:,0],X3[:,1],marker="o",facecolors="none",edgecolors="b")
```

Now let's do features normalization

```python
def featureNormalize(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    
    X_norm = (X - mu)/ sigma
    
    return X_norm, mu, sigma
```

Now computes eigenvectors of the covariance matrix of X

```python
def pca(X):
    m, n = X.shape[0], X.shape[1]
    
    sigma = 1/m * X.T @ X
    
    U, S, V = svd(sigma)
    
    return U, S, V
```

```python
X_norm, mu, std = featureNormalize(X3)
U, S = pca(X_norm)[:2]
```

### Visualizing PCA

```python
plt.scatter(X3[:,0],X3[:,1],marker="o",facecolors="none",edgecolors="b")
plt.plot([mu[0],(mu+1.5*S[0]*U[:,0].T)[0]],[mu[1],(mu+1.5*S[0]*U[:,0].T)[1]],color="black",linewidth=3)
plt.plot([mu[0],(mu+1.5*S[1]*U[:,1].T)[0]],[mu[1],(mu+1.5*S[1]*U[:,1].T)[1]],color="black",linewidth=3)
plt.xlim(-1,7)
plt.ylim(2,8)
```

```python
print("Top eigenvector U(:,1) =:", U[:,0])
```

Now let's compute the projection of the data

```python
def projectData(X, U, K):
    m = X.shape[0]
    U_reduced = U[:,:K]
    Z = np.zeros((m,K))
    
    for i in range(m):
        for j in range(K):
            Z[i,j] = X[i,:] @ U_reduced[:,j]
    
    return Z
```

```python
K=1
Z = projectData(X_norm, U, K)
print("Projection of the first example:",Z[0][0])
```

Now let's reconstruct our data again

```python
def recoverData(Z, U, K):
    m,n = Z.shape[0],U.shape[0]
    X_rec = np.zeros((m,n))
    U_reduced = U[:,:K]
    
    for i in range(m):
        X_rec[i,:] = Z[i,:] @ U_reduced.T
    
    return X_rec
```

```python
X_rec  = recoverData(Z, U, K)
print("Approximation of the first example:", X_rec[0,:])
```

Visualize our projection surface

```python
plt.scatter(X_norm[:,0],X_norm[:,1],marker="o",label="Original",facecolors="none",edgecolors="b",s=15)
plt.scatter(X_rec[:,0],X_rec[:,1],marker="o",label="Approximation",facecolors="none",edgecolors="r",s=15)
plt.title("The Normalized and Projected Data after PCA")
plt.legend()
```
